```python
from autoscraper import AutoScraper
```

```python
Liquidators_url="https://www.mca.gov.in/content/mca/global/en/contact-us/official-liquidators.html"
wanted_list=["Sh Vyomesh R. Sheth","Official Liquidator, Ministry of Corporate Affairs.","0731-2710051","ol[dot]indore[at]mca[dot]gov[dot]in","1st Floor, Old CIABuilding, Opp. GPO Comps Residency Area, Indore - 452001, Madhya Pradesh"]
```

```python
scraper=AutoScraper()
result=scraper.build(Liquidators_url,wanted_list)
Data = scraper.get_result_similar(Liquidators_url,grouped=True)
print(result)
```

```python
key = list(Data.keys())
print(key)
```

```python
scraper.set_rule_aliases({'rule_gocr':'Name','rule_2ibz':'Designation',"rule_m3j2":"Contact Number","rule_9o5m":"Contact Details"})
```

```python
scraper.keep_rules(['rule_gocr','rule_2ibz','rule_m3j2','rule_9o5m'])
scraper.save('Official Liquidators')
```
